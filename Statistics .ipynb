```python
import numpy as np
from scipy import stats

# Generate a list of 500 integers between 90 and 130
int_list = np.random.randint(90, 131, size=500)

# (i) Mean and Median
def calculate_mean(numbers):
    return np.mean(numbers)

def calculate_median(numbers):
    return np.median(numbers)

# (ii) Mode
def calculate_mode(numbers):
    mode_result = stats.mode(numbers)
    return mode_result.mode[0], mode_result.count[0]

# (iii) Weighted Mean
def calculate_weighted_mean(values, weights):
    return np.average(values, weights=weights)

# (iv) Geometric Mean
def calculate_geometric_mean(numbers):
    return stats.gmean(numbers)

# (v) Harmonic Mean
def calculate_harmonic_mean(numbers):
    return stats.hmean(numbers)

# (vi) Midrange
def calculate_midrange(numbers):
    return (np.min(numbers) + np.max(numbers)) / 2

# (vii) Trimmed Mean
def calculate_trimmed_mean(numbers, proportion_to_trim):
    return stats.trim_mean(numbers, proportion_to_trim)

# Example usage:
mean_value = calculate_mean(int_list)
median_value = calculate_median(int_list)
mode_value, mode_count = calculate_mode(int_list)

# For weighted mean, you need to provide weights
values = np.random.randint(90, 131, size=500)  # Example values
weights = np.random.rand(500)  # Example weights
weighted_mean_value = calculate_weighted_mean(values, weights)

geometric_mean_value = calculate_geometric_mean(int_list[int_list > 0])  # Geometric mean requires positive numbers
harmonic_mean_value = calculate_harmonic_mean(int_list[int_list > 0])  # Harmonic mean requires positive numbers
midrange_value = calculate_midrange(int_list)

# Trimmed mean, excluding 10% of the lowest and highest values
trimmed_mean_value = calculate_trimmed_mean(int_list, proportion_to_trim=0.1)

print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode: {mode_value}, Count: {mode_count}")
print(f"Weighted Mean: {weighted_mean_value}")
print(f"Geometric Mean: {geometric_mean_value}")
print(f"Harmonic Mean: {harmonic_mean_value}")
print(f"Midrange: {midrange_value}")
print(f"Trimmed Mean: {trimmed_mean_value}")

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Generate a list of 500 integers between 200 and 300
int_list2 = np.random.randint(200, 301, size=500)

# (i) Visualization
plt.figure(figsize=(14, 7))

plt.subplot(3, 1, 1)
sns.histplot(int_list2, kde=False, stat='density', bins=30, color='blue', alpha=0.7)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(int_list2), np.std(int_list2))
plt.plot(x, p, 'k', linewidth=2)
plt.title('Frequency Histogram & Gaussian Distribution')
plt.xlabel('Value')
plt.ylabel('Density')

plt.subplot(3, 1, 2)
sns.kdeplot(int_list2, fill=True, color='green')
plt.title('Frequency Smoothened KDE Plot')
plt.xlabel('Value')
plt.ylabel('Density')

plt.subplot(3, 1, 3)
sns.histplot(int_list2, kde=True, stat='density', bins=30, color='blue', alpha=0.7)
plt.title('Gaussian Distribution & Smoothened KDE Plot')
plt.xlabel('Value')
plt.ylabel('Density')

plt.tight_layout()
plt.show()

# (ii) Calculate Range
def calculate_range(numbers):
    return np.max(numbers) - np.min(numbers)

# (iii) Variance and Standard Deviation
def calculate_variance_and_std(numbers):
    variance = np.var(numbers, ddof=1)
    std_dev = np.std(numbers, ddof=1)
    return variance, std_dev

# (iv) Interquartile Range (IQR)
def calculate_iqr(numbers):
    q75, q25 = np.percentile(numbers, [75, 25])
    return q75 - q25

# (v) Coefficient of Variation
def calculate_coefficient_of_variation(numbers):
    mean = np.mean(numbers)
    std_dev = np.std(numbers, ddof=1)
    return (std_dev / mean) * 100

# (vi) Mean Absolute Deviation (MAD)
def calculate_mad(numbers):
    mean = np.mean(numbers)
    return np.mean(np.abs(numbers - mean))

# (vii) Quartile Deviation
def calculate_quartile_deviation(numbers):
    q75, q25 = np.percentile(numbers, [75, 25])
    return (q75 - q25) / 2

# (viii) Range-Based Coefficient of Dispersion
def calculate_range_based_coefficient_of_dispersion(numbers):
    range_value = np.max(numbers) - np.min(numbers)
    mean = np.mean(numbers)
    return (range_value / mean) * 100

# Example usage:
range_value = calculate_range(int_list2)
variance, std_dev = calculate_variance_and_std(int_list2)
iqr_value = calculate_iqr(int_list2)
cv_value = calculate_coefficient_of_variation(int_list2)
mad_value = calculate_mad(int_list2)
quartile_deviation_value = calculate_quartile_deviation(int_list2)
dispersion_value = calculate_range_based_coefficient_of_dispersion(int_list2)

print(f"Range: {range_value}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
print(f"IQR: {iqr_value}")
print(f"Coefficient of Variation: {cv_value}%")
print(f"MAD: {mad_value}")
print(f"Quartile Deviation: {quartile_deviation_value}")
print(f"Range-Based Coefficient of Dispersion: {dispersion_value}%")
```

```python
"""Write a Python class representing a discrete random variable with methods to calculate its expected
value and variance."""

class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
       
        if len(values) != len(probabilities):
            raise ValueError("The length of values and probabilities must be the same.")
        if not (0.999 <= sum(probabilities) <= 1.001):  # Allowing a small tolerance for floating-point arithmetic
            raise ValueError("The sum of probabilities must be 1.")
        
        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))

    def variance(self):
        
        mean = self.expected_value()
        return sum(prob * (value - mean) ** 2 for value, prob in zip(self.values, self.probabilities))

# Example usage
values = [1, 2, 3, 4]
probabilities = [0.1, 0.2, 0.3, 0.4]

random_variable = DiscreteRandomVariable(values, probabilities)

print(f"Expected Value: {random_variable.expected_value()}")
print(f"Variance: {random_variable.variance()}")

```

```python
"""Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and
variance of the outcomes."""

import numpy as np

def simulate_die_rolls(num_rolls):
    
    return np.random.randint(1, 7, size=num_rolls)

def calculate_expected_value(outcomes):
   
    return np.mean(outcomes)

def calculate_variance(outcomes):
    
    return np.var(outcomes, ddof=0)  # ddof=0 for population variance


num_rolls = 1000
outcomes = simulate_die_rolls(num_rolls)


expected_value = calculate_expected_value(outcomes)
variance = calculate_variance(outcomes)

print(f"Simulated Outcomes: {outcomes[:10]}...") 
print(f"Expected Value: {expected_value:.2f}")
print(f"Variance: {variance:.2f}")
```

```python
"""Create a Python function to generate random samples from a given probabilit distribution (e.g.,
binomial, Poisson) and calculate their mean and variance."""

import numpy as np

def generate_samples(distribution, params, num_samples):
   
    if distribution == 'binomial':
        n, p = params
        samples = np.random.binomial(n, p, num_samples)
    elif distribution == 'poisson':
        lam = params[0]
        samples = np.random.poisson(lam, num_samples)
    else:
        raise ValueError("Unsupported distribution. Please use 'binomial' or 'poisson'.")

    mean = np.mean(samples)
    variance = np.var(samples, ddof=0)  # Population variance

    return mean, variance


if __name__ == "__main__":
    
    n = 10  # Number of trials
    p = 0.5  # Probability of success
    num_samples = 1000
    binomial_mean, binomial_variance = generate_samples('binomial', (n, p), num_samples)
    print(f"Binomial Distribution - Mean: {binomial_mean:.2f}, Variance: {binomial_variance:.2f}")

   
    lam = 5  
    poisson_mean, poisson_variance = generate_samples('poisson', (lam,), num_samples)
    print(f"Poisson Distribution - Mean: {poisson_mean:.2f}, Variance: {poisson_variance:.2f}")
```

```python
"""Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute
the mean, variance, and standard deviation of the samples."""

import numpy as np

def generate_gaussian_samples(mean, std_dev, num_samples):
    
    
    samples = np.random.normal(mean, std_dev, num_samples)
    
   
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples, ddof=0)  
    sample_std_dev = np.sqrt(sample_variance)

    return sample_mean, sample_variance, sample_std_dev


if __name__ == "__main__":
    
    mean = 0          
    std_dev = 1       
    num_samples = 1000

    
    mean_value, variance_value, std_dev_value = generate_gaussian_samples(mean, std_dev, num_samples)
    
    
    print(f"Gaussian Distribution - Mean: {mean_value:.2f}")
    print(f"Variance: {variance_value:.2f}")
    print(f"Standard Deviation: {std_dev_value:.2f}")

```

```python
"""Write a Python function to calculate the probabilit density function (PDF) of a continuous random
variable for a given normal distribution."""

import numpy as np

def normal_pdf(x, mean, std_dev):
    

    variance = std_dev ** 2
    
   
    pdf = (1 / np.sqrt(2 * np.pi * variance)) * np.exp(-((x - mean) ** 2) / (2 * variance))
    
    return pdf


if __name__ == "__main__":
   
    mean = 0         
    std_dev = 1       
    x_value = 1.5     
    
    
    pdf_value = normal_pdf(x_value, mean, std_dev)
    
    
    print(f"PDF of normal distribution at x = {x_value}: {pdf_value:.4f}")

```
